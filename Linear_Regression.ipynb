#Importing required libraries


```python
import numpy as np
from matplotlib.pyplot import *
import pandas as pd
```

 Loading the txt file(converted to .txt from the given .csv) into 2 numpy arrays and plotting them

```python
x,y = np.loadtxt('points.txt', unpack = True)
X=x
plot(x,y, 'bo');
ylabel('Y-Axis')
xlabel('X-Axis')
grid()
show()
```

 Let us initialise the parameters

```python
m = np.size(y)
theta = np.zeros([2,1])
x = x.reshape([45,1])
x = np.hstack([x,np.ones_like(x)])
print(theta,'\n',m)
```

We reshaped x to 45 X 2 matrix for mathematical purpose,let's look at the features.

```python
print(x,'\n',y)
print(x.shape)
print(theta.shape)
```

Defining the cost function

```python
def computecost(x,y,theta):
    
    prefix = 1/(2*m)
    error = np.sum((y- (x@theta))**2)
    return prefix*error
```

Computing cost for initial theta

```python
print(computecost(x,y,theta))
```

Our goal is to reduce the value of cost function,
We shall use the gradient descent method

```python
def gradient(x,y,theta):
    
    alpha = 0.00000001
    iteration = 20
#gradient descent algorithm
    J_history = np.zeros([iteration, 1]);
    for iter in range(0,20):
        if(iter % 2 == 0):
          print("Epoch: #",iter//2)
          plot(X,y,'go');
          plot(X, x@theta ,'-');
          ylabel('Y-Axis')
          xlabel('X-Axis')
          grid()
          show()
        error = (np.dot(x,theta)) -y
        temp0 = theta[0] - ((alpha/m) * np.sum(error*x[:,0]))
        temp1 = theta[1] - ((alpha/m) * np.sum(error*x[:,1]))
        theta = np.array([temp0,temp1]).reshape(2,1)
        J_history[iter] = (1 / (2*m) ) * (np.sum(((np.dot(x,theta))-y)**2))   #compute J value for each iteration 
    return theta, J_history
```

Let us compute the regressed Parameters after passing Theta to the gradient descent function

```python
regressedTheta, J = gradient(x,y,theta)
print("Regressed Theta = \n" ,regressedTheta, "\n")
```

Finally, Plotting it!

```python
plot(X,y,'go');
plot(X, x@regressedTheta ,'-');
ylabel('Y-Axis')
xlabel('X-Axis')
grid()
show()
```
